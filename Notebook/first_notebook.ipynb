```python
df.drop(columns=["customerID"],inplace=True)
```

```python
df["TotalCharges"]=pd.to_numeric(df["TotalCharges"],errors="coerce")
```

```python
df["tenure"]=pd.cut(df["tenure"],bins=[0,12,24,48,100],labels=["High Chances","Chances","low Chances","very low chances"],include_lowest=True)
```

```python
df.isnull().sum()
```

```python
df["TotalCharges"].isnull().sum()/len(df["TotalCharges"])
```

```python
df["TotalCharges"].skew()
```

```python
df["InternetService"].unique()
df["InternetService_used"]=df["InternetService"].map({"No":"No","Fiber optic":"Yes","DSL":"Yes"})
```

```python
df["MultipleLines"].unique()
df["MultipleLines"]=df["MultipleLines"].map({"No phone service":"No","Yes":"Yes","No":"No"})
```

```python
service=["PhoneService","MultipleLines","InternetService_used","OnlineSecurity","OnlineBackup","DeviceProtection","TechSupport","StreamingTV","StreamingMovies"]
df["Services"]=df[service].eq("Yes").sum(axis=1)
```

```python
df["Services"]
```

```python
df[service]
```

```python
df["Churn"]=df["Churn"].map({"Yes":1,"No":0})
```

```python
df["risk"]=((df["tenure"].eq("High Chances")) & (df["Contract"]=="Month-to-month")).astype(int)
```

```python
df.groupby("tenure")["Contract"].value_counts()
```

```python
df.info()
```

```python
df["Partner"] = df["Partner"].str.strip().str.capitalize()
df["Dependents"] = df["Dependents"].str.strip().str.capitalize()
```

```python
df["Partner"] = df["Partner"].map({"Yes": 1, "No": 0})
df["Dependents"] = df["Dependents"].map({"Yes": 1, "No": 0})
df["family_present"]=(df["SeniorCitizen"]+df["Partner"]+df["Dependents"])
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
encode=["gender","tenure","InternetService","Contract","PaperlessBilling","PaymentMethod"]
encode=encode+service
print(encode)
```

```python
preprocessing=ColumnTransformer(transformers=[("encoding",OneHotEncoder(drop="first",handle_unknown="ignore"),encode),("impute",SimpleImputer(strategy="median"),["TotalCharges"])],remainder="passthrough")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score,recall_score,precision_score,classification_report,confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df.drop(columns=["Churn"]),df["Churn"],random_state=42,test_size=0.25,stratify=df["Churn"])
```

```python
import dagshub

dagshub.init(repo_owner='Mukul9112003', repo_name='project3_Telecom_Customer_Churn_Prediction_System', mlflow=True)
mlflow.set_experiment(experiment_name="handle imbalance")

with mlflow.start_run():
  pipeline=Pipeline(steps=[
    ("pre",preprocessing),
    ("model",LogisticRegression(class_weight="balanced"))
  ])
  pipeline.fit(X_train,Y_train)
  predict=pipeline.predict(X_test)
  accuracy=accuracy_score(Y_test,predict)
  precision=precision_score(Y_test,predict)
  recall=recall_score(Y_test,predict)
  confusion=confusion_matrix(Y_test,predict)
  report=classification_report(Y_test,predict)
  mlflow.log_metric('accuracy_score',accuracy)
  mlflow.log_metric('precision_score',precision)
  mlflow.log_metric('recall_score',recall)
```

```python
import numpy as np
import mlflow
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, precision_recall_curve
)
from xgboost import XGBClassifier

neg = (Y_train == 0).sum()
pos = (Y_train == 1).sum()
scale_pos_weight = neg / pos if pos != 0 else 1

with mlflow.start_run():
    pipeline = Pipeline(steps=[
        ("pre", preprocessing),
        ("model", XGBClassifier(
            eval_metric="logloss",
            scale_pos_weight=scale_pos_weight,
            use_label_encoder=False
        ))
    ])

    param_grid = {
        "model__n_estimators": [100, 200],
        "model__max_depth": [3, 5],
        "model__learning_rate": [0.01, 0.1]
    }

    grid = GridSearchCV(
        estimator=pipeline,
        param_grid=param_grid,
        cv=5,
        n_jobs=-1,
        scoring="f1"   
    )

    grid.fit(X_train, Y_train)
    best = grid.best_estimator_

    probs = best.predict_proba(X_test)[:, 1]

    thresholds = np.arange(0.1, 0.9, 0.01)

    best_thresh = 0.5
    best_f1 = 0

    for t in thresholds:
        preds = (probs > t).astype(int)
        f1 = f1_score(Y_test, preds)

        if f1 > best_f1:
            best_f1 = f1
            best_thresh = t

    print("Best Threshold:", best_thresh)
    print("Best F1:", best_f1)

    final_pred = (probs > best_thresh).astype(int)

    accuracy = accuracy_score(Y_test, final_pred)
    precision = precision_score(Y_test, final_pred)
    recall = recall_score(Y_test, final_pred)
    auc = roc_auc_score(Y_test, probs)

    mlflow.log_metric("accuracy", accuracy)
    mlflow.log_metric("precision", precision)
    mlflow.log_metric("recall", recall)
    mlflow.log_metric("f1", best_f1)
    mlflow.log_metric("roc_auc", auc)
    mlflow.log_metric("best_cv_score", grid.best_score_)

    mlflow.log_param("threshold", best_thresh)
    mlflow.log_params(grid.best_params_)

    mlflow.sklearn.log_model(best, "model")
```
